# Network Graphs
A simple example using Brunel for Graph Layout.

```python
import pandas as pd
import brunel

nodes = pd.read_csv("data/sample_nodes.csv")
edges = pd.read_csv("data/sample_edges.csv")
```

```python
%%brunel data('edges') key(From,To) edge + 
         data('nodes') network color(Node) key(Node) legends(none) style('fill-opacity:1;size:20px') label(Node)
```
