# Network Graph Example
A simple example of a network graph layout

```r
library(brunel)
nodes <- read.csv('data/sample_nodes.csv', header=TRUE)
edges <- read.csv('data/sample_edges.csv', header=TRUE)
```

```r
brunel ("data('edges') key(From,To) edge + data('nodes') network color(Node) key(Node) legends(none) style('fill-opacity:1;size:20px') label(Node)")
```
